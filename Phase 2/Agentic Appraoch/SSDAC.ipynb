```python
import requests
from bs4 import BeautifulSoup
import time
import re
import random
from langchain.llms import Ollama
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import http.client
import json



def google_search(query):
    conn = http.client.HTTPSConnection("google.serper.dev")
    payload = json.dumps({
      "q": query
    })
    headers = {
      'X-API-KEY': '46c876d4cd57f94fc82e98a74e64edfa757c744c',
      'Content-Type': 'application/json'
    }
    conn.request("POST", "/search", payload, headers)
    res = conn.getresponse()
    data = res.read()
    return data.decode("utf-8")



model = Ollama(model="llama3.1:8b-instruct-q8_0", temperature=0)

prompt_template = PromptTemplate(
    input_variables=["description", "categories"],
    template=(
        """
       Product description:
{description}

What is the most specific and accurate product category that perfectly fits the description?

Please carefully evaluate each category, and **choose only one exact category name** from the list below without modifying or generating a new one. If a more specific category exists that matches closely, choose that over a broader category.

Here are the possible categories (each on a new line):
{categories}

Provide only the exact product category string from the list above, with no modifications or rewording.
       """
    )
)

chain = LLMChain(llm=model, prompt=prompt_template)

def clean_llm_output(output):
    pattern = r"[\"',()]+|\s{2,}"  # Matches quotes, commas, brackets, and excessive whitespace
    cleaned_output = re.sub(pattern, ' ', output)
    cleaned_output = ' '.join(cleaned_output.split())
    cleaned_output = cleaned_output.lower()
    return cleaned_output

def product_pipeline(product_queries):
    results = []
    
    
    # Use tqdm to create a progress bar, with total being the length of product_queries
    for product_query in product_queries:
        # Get product description from Google search
        product_description = google_search(product_query)
        
        if product_description:
            # Run the LLM chain with the product description
            temp_dict = {}
            output = chain.run(description=product_description, categories=supergroup)
            cleaned_output = clean_llm_output(output)
            temp_dict["supergroup"] = cleaned_output
            group_range = supergroup_dict.get(cleaned_output, [])
            
            if not group_range:
                temp_dict["group"] = "na"
                temp_dict["module"] = "na"
                results.append(temp_dict)
                continue
            elif len(group_range) == 1:
                group_ans = group_range[0]
            else:
                groups = "\n".join(group_range)
                group_output = chain.run(description=product_description, categories=groups)
                group_ans = clean_llm_output(group_output)
                
            temp_dict["group"] = group_ans    
            module_range = group_dict.get(group_ans, [])
            
            if not module_range:
                module_ans = "na"
            elif len(module_range) == 1:
                module_ans = module_range[0]
            else:
                modules = "\n".join(module_range)
                module_output = chain.run(description=product_description, categories=modules)
                module_ans = clean_llm_output(module_output)
                
            temp_dict["module"] = module_ans
            results.append(temp_dict)
        else:
            print(product_query)
            results.append({"supergroup":"na","group":"na","module":"na"})
    return results

t1 = time.time()
# Example of running the pipeline with multiple products
product_queries = ["cp irr pld pork","2 stroke oil 500 ml","bpk pop tarts chocotastic 2","pa nat btl 125 ml","d brun optns botti"]

# This will now show a progress bar as it processes each product
category_results = product_pipeline(product_queries)
category_results
```

```python
import time
from torch.profiler import profile, record_function, ProfilerActivity

def predict(text):
    start = time.time()
    with profile(activities=[ProfilerActivity.CPU, ProfilerActivity.CUDA],with_flops=True) as prof:
        category_results = product_pipeline([text])
    print("Inference time :"+str(time.time()-start))
    #print(prof.key_averages().table(sort_by="flops",row_limit=10))
    print("GFLOPs during testing") #GigaFLOPs
    print(sum(k.flops for k in prof.key_averages())/1e9)
        
    return category_results
# Example prediction
sample_text = "14 in hybrid blade"
predictions = predict(sample_text)
print(predictions)
```
